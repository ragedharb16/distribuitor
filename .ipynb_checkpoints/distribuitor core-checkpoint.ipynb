```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df= pd.read_csv("\\Users\\raged\\Downloads\\data (8).csv")
```

```python
df
```

```python
#defining a function to plot the probability graph

def probability(col):
    col = col
    feature =df[col]
    mean = feature.mean()
    median =feature.median()
    std = feature.std()
    plus_std = mean + std
    minus_std = mean - std
    kurt = round(stats.kurtosis(feature, fisher =False),1)

    fig, ax =plt.subplots(figsize =(10,6))

    sns.histplot(feature, ax=ax, stat ="probability" ,kde =True)
    ax.axvline(mean, color ="red", label = f'Mean ={mean:,.2f}')
    ax.axvline(median, color ="green", label =f'Median = {median:,.2f}')
    ax.axvline(plus_std, color ="black",ls ="--", label =f'  + 1 std = {plus_std:,.2f}')
    ax.axvline(minus_std, color ="black", ls ="--", label =f'  - 1 std = {minus_std:,.2f}')
    ax.axvspan(plus_std, minus_std, color ="yellow", zorder =0)
    ax.set_title(f'The probability of {col}')
    ax.legend()
    
    print(f'Kurtosis is: {kurt}')
```

```python
probability("VitD_levels")
```
